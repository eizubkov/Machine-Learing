```python
from math import sin, exp
import  matplotlib.pyplot  as  plt
from scipy.optimize import minimize
import numpy as np
```

```python
%matplotlib inline
```

```python
f = lambda x: sin(float(x) / 5) * exp(float(x) / 10) + 5. * exp(float(-x) / 2)
```

```python
x = np.arange(1, 30, 0.01)
y = [f(i) for i in x]
fixg, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, y)
```

```python
f_min = minimize(f, 2, method='BFGS')
f_min = f_min.fun
round(f_min, 2)
```

```python
f_min = minimize(f, 30, method='BFGS')
f_min = f_min.fun
round(f_min, 2)
```

```python
from scipy.optimize import differential_evolution
```

```python
bounds = [(1, 30)]
result = differential_evolution(f, bounds)
round(result.fun, 2)
```

```python
h = lambda x: int(f(x))
```

```python
x = np.arange(1, 30, 0.01)
y = [h(i) for i in x]
fixg, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, y)
```

```python
h_min = minimize(h, 30, method='BFGS')
h_min.fun
```

```python
h_min = differential_evolution(h, bounds)
h_min.fun
```
