```python
from math import sin, exp
import  matplotlib.pyplot  as  plt
from scipy.optimize import minimize, golden
import numpy as np
```

```python
%matplotlib inline
```

```python
f = lambda x: sin(float(x) / 5) * exp(float(x) / 10) + 5. * exp(float(-x) / 2)
```

```python
x = np.arange(1, 30, 0.01)
y = [f(i) for i in x]
fixg, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, y)
```

```python
f_min = minimize(f, 2, method='BFGS')
f_min = f_min.fun
round(f_min, 2)
```

```python
f_min = minimize(f, 30, method='BFGS')
f_min = f_min.fun
round(f_min, 2)
```
